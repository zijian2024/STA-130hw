```python
1，

Predictor variables: This is the variable used to predict changes in the outcome variable. It is typically denoted as \( X \) and represents values that are known or observed.

Outcome variables: This is the variable we want to predict. It is usually denoted by \( Y \). It is the variable whose variations we seek to model based on changes in \( X \).

Intercept (\( \beta_0 \)): This constant term represents the expected value of \( Y \) when \( X = 0 \). It is the starting point of the regression line on the \( Y \)-axis.

Slope (\( \beta_1 \)): The slope coefficient indicates the expected change in \( Y \) for a one-unit increase in \( X \). It reflects the direction and strength of the relationship between \( X \) and \( Y \). If \( \beta_1 \) is positive, an increase in \( X \) is associated with an increase in \( Y \). If \( \beta_1 \) is negative, an increase in \( X \) corresponds to a decrease in \( Y \).

Error Term (\( \epsilon \)): The error term represents the deviation of observed \( Y \) values from the values predicted by the model. This term accounts for the fact that the linear model does not capture all the variability in \( Y \). In a standard linear regression model, \( \epsilon \) is assumed to follow a normal distribution with a mean of 0 and a constant variance \( \sigma^2 \), symbolized as \( \epsilon \sim N(0, \sigma^2) \).

https://chatgpt.com/share/672be72d-f614-800b-b65f-3ce3d1782ba9
    
```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# 模拟数据
np.random.seed(0)
X = np.linspace(0, 10, 100)
# 假设 β0 = 2，β1 = 3，添加误差项 ϵ
β0 = 2
β1 = 3
ϵ = np.random.normal(0, 1, size=X.shape)
Y = β0 + β1 * X + ϵ

# 线性回归拟合
slope, intercept, r_value, p_value, std_err = stats.linregress(X, Y)

# 绘制数据点和回归线
plt.scatter(X, Y, color='blue', label='Data')
plt.plot(X, intercept + slope * X, color='red', label='Fitted line')
plt.xlabel('X (Predictor)')
plt.ylabel('Y (Outcome)')
plt.legend()
plt.title('Simple Linear Regression')
plt.show()

# 输出拟合结果
print(f'Intercept (β0): {intercept:.2f}')
print(f'Slope (β1): {slope:.2f}')
print(f'R-squared: {r_value**2:.2f}')
```

```python
4
The theoretical line (orange dashed line) represents the regression line generated by the known true parameters (β0 and β1), representing the prediction under ideal conditions.
The fitted line (solid red line) is the regression line that is fitted using the least square method based on the sample data. It is affected by random error terms in the data, so it may deviate somewhat from the theoretical line.
The difference between the two: The theoretical line assumes no random error at all, while the fitted line reflects random sampling variations and errors in the sample data, so it is closer to the actual observations. This illustrates how variation in the sample data affects the fit results of the model.
```

```python
5
Here's a detailed explanation for Question 4 on how `fitted_model.fittedvalues` are derived:

### Explanation of How `fitted_model.fittedvalues` Are Derived:
In a simple linear regression model, the fitted values are the predicted outcomes calculated using the estimated coefficients from the model. The fitted values, denoted as \( \hat{Y}_i \), are calculated using the following equation:

\[
\hat{Y}_i = \hat{\beta}_0 + \hat{\beta}_1 x_i
\]

#### Definitions:
- **\( \hat{Y}_i \)**: The predicted value for the \( i \)-th observation.
- **\( \hat{\beta}_0 \) and \( \hat{\beta}_1 \)**: The estimated coefficients (intercept and slope) obtained from the model. These can be found in `fitted_model.params` or `fitted_model.params.values`. These coefficients are estimated by minimizing the sum of squared residuals between the observed \( Y \) values and the predicted \( \hat{Y} \) values.
- **\( x_i \)**: The predictor value for the \( i \)-th observation from the dataset.

### Derivation Using `fitted_model.summary().tables[1]` for Estimated Coefficients:
`fitted_model.summary().tables[1]` provides a summary table showing the estimated intercept (\(\hat{\beta}_0\)) and slope (\(\hat{\beta}_1\)) along with their standard errors and p-values. These estimated coefficients can be directly extracted using `fitted_model.params`.

### Calculating Fitted Values:
The fitted values (`fitted_model.fittedvalues`) are calculated by applying the estimated coefficients to the predictor values (\( x_i \)) in the dataset:

\[
\text{fitted values} = \hat{\beta}_0 + \hat{\beta}_1 \cdot x_i
\]

This equation aligns with the structure of the theoretical model but uses the estimated parameters from the sample data rather than the true population parameters. In essence, `fitted_model.fittedvalues` is a direct application of the linear equation, using the estimated intercept and slope from the fitted model for each predictor value. This produces a series of predicted outcomes that best fit the observed data, following the ordinary least squares (OLS) method.

### Summary:
`fitted_model.fittedvalues` represents the predicted values in a linear regression model, calculated using the estimated intercept (\(\hat{\beta}_0\)) and slope (\(\hat{\beta}_1\)) from `fitted_model.params`.

The equation:

\[
\hat{Y}_i = \hat{\beta}_0 + \hat{\beta}_1 x_i
\]

ap
```

```python
- ** View the p value ** :
- If p is \(p &gt; 0.05 \), there is insufficient evidence to reject the null hypothesis, indicating that there is insufficient evidence for a linear relationship between the waiting time and the eruption duration in the average case.
- If p has a value \(p \leq 0.05 \), then the null hypothesis can be rejected, stating that in the average case, there is sufficient evidence for a linear relationship between the waiting time and the eruption duration.

- ** How to interpret the results ** :
- ** Meaning of the P-value ** : The p-value represents the probability of observing the current sample result or a more extreme result if \(H_0 \) is assumed to be true.
- ** Acceptance and Rejection of null hypothesis ** :
- ** High p value ** (e.g. \(p &gt; 0.1 \)) : indicates that there is insufficient evidence against the null hypothesis.
- ** Low p values ** (e.g. \(p \leq 0.05 \)) : indicates that there is sufficient evidence to reject the null hypothesis, indicating that a significant linear relationship exists.

### Summary:
1. ** Null hypothesis \(H_0 \)** : There is no linear relationship between the wait time and the eruption duration (\ (\beta_1 = 0\)).
2. ** Alternative hypothesis \(H_a \)** : There is a linear relationship between the waiting time and the eruption duration (\ (\beta_1 \neq 0\)).
3. ** Interpretation results ** : Judging whether the null hypothesis can be rejected according to the P-value of the linear regression model, and interpreting the evidence in the data relative to the null hypothesis.
 
```

```python
import numpy as np
from scipy import stats


long_wait_limit = 71 
long_wait = old_faithful.waiting > long_wait_limit


bootstrapped_slopes = []
for _ in range(1000):  
    sample = old_faithful[long_wait].sample(n=len(old_faithful[long_wait]), replace=True)
    model_boot = smf.ols('duration ~ waiting', data=sample).fit()
    bootstrapped_slopes.append(model_boot.params['waiting'])


simulated_slopes = []
old_faithful_simulation = old_faithful[long_wait].copy()
old_faithful_simulation['duration'] = 1.65 + 0 * old_faithful_simulation['waiting'] + stats.norm(loc=0, scale=0.37).rvs(size=len(old_faithful_simulation))

for _ in range(1000):  
    model_sim = smf.ols('duration ~ waiting', data=old_faithful_simulation).fit()
    simulated_slopes.append(model_sim.params['waiting'])


confidence_interval = np.percentile(bootstrapped_slopes, [2.5, 97.5])
print("95% Bootstrap Confidence Interval for Slope Coefficient:", confidence_interval)
print("Does the confidence interval contain 0?", 0 in range(int(confidence_interval[0]), int(confidence_interval[1])))

```

```python
import numpy as np
import pandas as pd


import seaborn as sns
old_faithful = sns.load_dataset('geyser')
old_faithful['kind'] = old_faithful['waiting'].apply(lambda x: 'short' if x < 68 else 'long')


observed_diff = old_faithful.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]


perm_diffs = []
for _ in range(1000):
    shuffled_kind = old_faithful['kind'].sample(frac=1, replace=False).values
    permuted_diff = old_faithful.assign(kind_shuffled=shuffled_kind).groupby('kind_shuffled')['duration'].mean().iloc[::-1].diff().values[1]
    perm_diffs.append(permuted_diff)

# 计算 p 值
p_value = np.mean(np.abs(perm_diffs) >= np.abs(observed_diff))
print(f'Observed difference: {observed_diff}')
print(f'Permutation test p-value: {p_value}')
```

```python
# 引导法
bootstrap_diffs = []
for _ in range(1000):
    boot_sample = old_faithful.groupby('kind').apply(lambda x: x.sample(n=len(x), replace=True)).reset_index(drop=True)
    boot_diff = boot_sample.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]
    bootstrap_diffs.append(boot_diff)

ci_lower, ci_upper = np.quantile(bootstrap_diffs, [0.025, 0.975])
print(f'95% Bootstrap Confidence Interval: ({ci_lower}, {ci_upper})')
```

```python
(a) Explanation of the working principle of the two simulation methods:
Substitution test: By randomly shuffling labels, a distribution that conforms to the null hypothesis is generated to evaluate the significance of the sample statistic.
Guide method: By sampling with retractions, generate multiple samples, estimate the distribution of statistics and calculate confidence intervals.
(b) Comparison with the indicator variable model in question 11:
Similarity: Both substitution test and guide method are used to evaluate whether there is a significant difference in eruption duration between short and long wait times, which is consistent with the purpose of testing the difference by model in question 11.
The difference: The index variable model in Question 11 is tested using regression coefficients and p-values of linear regression, while the replacement test and guide method are non-parametric methods that do not rely on normality or linear assumptions.
```
